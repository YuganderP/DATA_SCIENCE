```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
x
y=x**2
```

```python
y
```

```python
plt.plot(x,y) # plots a line graph
plt.title("Title of the graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
#subplots
```

```python
plt.subplot(2,2,1) # row column position  
plt.plot(x**2,y**2)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
# object oriented way : to make changes to the canvas
fig=plt.figure(figsize=(6,8)) #(pixles) 
axis1=fig.add_axes([0.1, 0.1, 1, 1]) #[left distance ,bottom distance, width , height]
axis1.plot(x,y)
axis2=fig.add_axes([0.8, 0.9, 0.1, 0.1]) #[left distance ,bottom distance, width , height]
axis2.plot(x,y)
```
