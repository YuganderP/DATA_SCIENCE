# iterating in Numpy


```python
a=np.array([1,2,3,4,5,6,7,8])
for i in a:
    print(i)
```

```python
b=np.array([[1,2,3,4],[5,6,7,8]])
for i in np.nditer(b,flags=['buffered'],op_dtypes=['S']):
    print(i)
```
