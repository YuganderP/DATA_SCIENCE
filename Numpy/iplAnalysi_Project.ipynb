```python
# which team won most matches

df.head()
```

```python
match_wins=df['match_winner'].value_counts()
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title("most matches won by a team ")
```

```python
# toss decision trends
df.head()
```

```python
data=df.groupby('toss_winner')['toss_decision'].value_counts()
```

```python
sns.countplot(x=df['toss_decision'])
plt.title("toss decision trends")
```

```python
# toss winner vs match winner

count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count/df['match_id'].count())*100
print(percentage.round(2))
```

```python
# how do teams win runs vs wickets
df['won_by'].value_counts()
sns.countplot(x=df['won_by'])
plt.title('runs vs wickets')
```

```python
df.groupby('match_winner')['won_by'].value_counts()
```

```python
# most player of the match awards
count=df['player_of_the_match'].value_counts().head(10)
```

```python
sns.barplot(x=count.values,y=count.index)
```

```python
# top scores
```

```python
count=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(5)
```

```python
count.plot(kind='bar')
```

```python
count=df.groupby('best_bowling')['best_bowling_figure'].count().sort_values(ascending=False).head(10)
count
```

```python
count.plot(kind='bar')
```
