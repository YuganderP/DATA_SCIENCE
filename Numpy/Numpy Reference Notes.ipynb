```python
# different dimension arrays
oneD=np.array([1,2,3,4,5])
twoD=np.array([[1,2,3],[4,5,6]])
threeD=np.array([[[1,2],[2,3],[4,5]]])
print(oneD)
print(oneD.ndim)
print(twoD)
print(twoD.ndim)
print(threeD)
print(threeD.ndim)
```

```python
# array filled with zeros
zeros_1D=np.zeros(4)
print(zeros_1)
```

```python
zeros_2D=np.zeros((3,4))
print(zeros_2D)
```

```python
#arrays filled with one 
ones_1D=np.ones(4)
print(ones_1D)
```

```python
ones_2d=np.ones((3,4))
print(ones_2d)
```

```python
empty_2d=np.empty((3,4))
print(empty_2d)
```

```python
# fill elements 

arange_1=np.arange(4)
print(arange_1)
```

```python
arange_2=np.arange(3,8)
print(arange_2)
```

```python
# array with diagonal elements 

diagonal_1=np.eye(3)
print(diagonal_1)
```

```python
diagonal_2=np.eye(3,5)
print(diagonal_2)
```

```python
#evenly spaced array

ar_lin=np.linspace(0,10,5)
print(ar_lin)
```
