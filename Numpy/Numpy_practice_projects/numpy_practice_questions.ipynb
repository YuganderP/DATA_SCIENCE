## 7. Recommendation System Mini Project

Create a user-movie rating matrix.

Tasks:
- Average movie rating
- Most liked movie
- Find similar users
- Recommend movies


### Hints
- Think of rows as users and columns as movies
- Use dot products for similarity
- Use averages across columns
- Broadcasting may help


```python
# Write your solution here

users= np.arange(1,6).reshape(5,1)

movies=np.random.randint(0,5,30).reshape(5,6)

data=np.concatenate((users,movies),axis=1)
print(data.shape)
print()

avg_movie_rating=np.mean(data[:,1:],axis=0)
print(avg_movie_rating)
most_liked_movie = np.argmax(avg_movie_rating)
print(most_liked_movie+1) # gives most liked movie
print()

similar_users=np.array([])
print(similar_users)
print()
# related users 
for user in range(0,data.shape[0]):
    for movies in range(0,data.shape[0]):
        if(user==movies):
            continue
        a=np.dot(data[user,1:],data[movies,1:])
        similar_users=np.append(similar_users,a)


print()

similar_users=(similar_users).reshape(5,4)
print(similar_users)
print()
print()

# now we will get the list of similar users 

most_similar_users= []

print(similar_users.shape)
for i in range(0,similar_users.shape[0]):

    # find best similarity position in current row
    b = np.argmax(similar_users[i,:])

    # mapping actual user ids because self-user was skipped
    compared_users = [x for x in range(data.shape[0]) if x != i]

    # actual most similar user
    actual_user = compared_users[b]

    most_similar_users.append((i, actual_user))


# recommend movies

recommendations = []

for i in most_similar_users:

    current_user = i[0]
    similar_user = i[1]

    # current user movie ratings
    current_user_movies = data[current_user,1:]

    # similar user movie ratings
    similar_user_movies = data[similar_user,1:]

    # movies not watched by current user
    unwatched_movies = (current_user_movies == 0)

    # recommend movies liked by similar user
    recommended = np.where(
        (unwatched_movies) & (similar_user_movies >= 4)
    )[0]

    recommendations.append((current_user, recommended))

print(recommendations)
```
