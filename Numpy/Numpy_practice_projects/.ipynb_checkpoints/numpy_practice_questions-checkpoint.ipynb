## 1. Student Marks Analyzer

Create a 2D NumPy array where:
- rows = students
- columns = subjects

Tasks:
- Find average marks per student
- Find highest scorer
- Find subject-wise averages
- Find failed students (<35)
- Find topper per subject


### Hints
- Use `np.mean(..., axis=?)`
- Use `np.max()` and `np.argmax()`
- Think carefully about `axis=0` vs `axis=1`
- Boolean filtering will help for failed students


```python
# Write your solution here
import numpy as np
student_data=np.array([[1,90,80,90,80,80,99],[2,50,80,40,80,80,80],[3,50,20,90,80,40,90],[4,30,80,90,50,20,90],[5,10,40,50,20,11,70],[6,80,45,36,45,36,34],[7,66,84,45,57,26,46],[8,46,46,35,35,56,35]])
print(student_data)
print("student wise avg")
print(np.mean((student_data[:,1:]),axis=1))
print("top scorer")
a=(np.argmax(np.mean((student_data[:,1:]),axis=1)))
print(student_data[a,0])
print()
print("subject wise average")
print(np.mean(student_data[:,1:],axis=0))
print()
print("failed students")
print(student_data[np.mean((student_data[:,1:]),axis=1)<35])
print(np.mean((student_data[:,1:]),axis=1)<35)

```

## 2. Weather Temperature Analysis

Generate random temperatures for 30 days.

Tasks:
- Find hottest and coldest day
- Find average temperature
- Find days above average
- Reshape into weekly data
- Find weekly averages


### Hints
- Use `np.random.randint()`
- Use `reshape()` for weekly grouping
- Use boolean indexing for above-average days
- `np.mean()` will help for averages


```python
# Write your solution here

weather_data=np.random.randint(20,40,30)
max_temp=np.max(weather_data)
min_temp=np.min(weather_data)
avg_temp=np.mean(weather_data)
days_above_avg_temp=(weather_data>avg_temp)

print(weather_data[days_above_avg_temp])
weather_data=np.append(weather_data,[0,0,0,0,0])
weather_data=(weather_data.reshape(5,7))
weekly_avg= np.mean(weather_data[0:4,:])
print(weather_data)
print()

print(max_temp,min_temp)
print()
print(avg_temp)
print()
print(weekly_avg)
print()

```

## 3. Image Pixel Simulator

Create a random 256x256 grayscale image array.

Tasks:
- Increase brightness
- Decrease brightness
- Crop image
- Flip image horizontally
- Normalize pixel values


### Hints
- Pixel values are usually between 0–255
- Add/subtract values for brightness changes
- Use slicing for cropping
- Use `np.flip()` or slicing tricks
- Normalize using division


```python
# Write your solution here
image = np.random.randint(0,255,size=(256,256))
print(image)
print("-------------------")
# increase bright ness
print(image+1)
print("-------------------")
# decrease bright ness 
print(image-1)
# cropping the image 
print("-------------------")
#cropping the image
print(image[250:,250:])
print("-------------------")
# flipping the image 
print(np.flip(image))
print("-------------------")
#normalizing the image 

print((image-np.min(image))/(np.max(image)-np.min(image)))



```

## 4. Sales Data Dashboard

Create monthly sales data for multiple products.

Tasks:
- Total sales
- Best-selling product
- Monthly growth
- Highest sales month
- Cumulative sales


### Hints
- Use `axis` carefully
- Use `np.sum()`
- Use `np.cumsum()`
- Use subtraction to calculate growth


```python
# Write your solution here
# the data is is 10X2  product id followed by sales 
products = np.arange(100,110).reshape(10,1)
sales = np.random.randint(10,432,size=(10,30))
data=np.concatenate((products,sales),axis=1)
# print(data)
print()
monthly_sales = np.sum(data[0:,1:],axis=1)
print(data[np.argmax(monthly_sales),0])
print()
print()
highest_sales_month= np.sum(data[:,1:],axis=1)
print(highest_sales_month)




```
