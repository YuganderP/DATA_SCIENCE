```python
import pandas as pd
```

```python
#series data structure 1-Dimensional
x=[3,4,5,6,7,8]
var=pd.Series(x)
print(var)
```

```python
print(type(var))
```

```python
#using custom index and desired datatype
# and defining name
x=[1,2,3,4]

var= pd.Series(x,index=['a','b','c','d'],dtype='float',name="python")
print(var)
```

```python
dic={"name":['a','b','c','d'],'pop':[12,13,14,15],'rank':[1,4,3,2]}
var_1=pd.Series(dic)
print(var_1)
```
