# Arthematic Operations in pandas

```python
# we will perform arthematic operations on the column data 
dic={'a':[1,2,3,4],'b':[3,4,5,6]}
df_3=pd.DataFrame(dic)
```

```python
print(df_3)
```

```python
# df_3['c']=df_3['a']+df['b']
# print(df_3)

df_3['c']=(df_3['a']+df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']-df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']*df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']/df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']//df_3['b'])
print(df_3)
print()
```

```python
df_3['c']=((df_3['a']<3) & (df_3['b']<5))
print(df_3)
print()
```

# Delete and Insert Data


```python
dic={'A':[1,2,3,4,5],'B':[6,7,8,9,10]}

var=pd.DataFrame(dic)

var.insert()
```

```python
# var.insert(index,"column name",[data])
```

```python
var.insert(2,'c',[10,11,12,13,14])
print(var)
```

```python
# copying data from existing table

var['d']=var['A'][:3]
print(var)
```

```python
# deleting data
var['d']=var['A'][:3]

var.pop('d')
print(var)
```
