```python
import pandas as pd
```

```python
#series data structure 1-Dimensional
x=[3,4,5,6,7,8]
var=pd.Series(x)
print(var)
```

```python
print(type(var))
```

```python
#using custom index and desired datatype
# and defining name
x=[1,2,3,4]

var= pd.Series(x,index=['a','b','c','d'],dtype='float',name="python")
print(var)
```

```python
dic={"name":['a','b','c','d'],'pop':[12,13,14,15],'rank':[1,4,3,2]}
var_1=pd.Series(dic)
print(var_1)
```

```python
var_2= pd.Series(12)
print(var_2)
print()
var_2=pd.Series(12,index=[1,2,3,4,5,6])
print(var_2)


```

```python
# there is is mismatch of rows in numpy we will get broadcasting error
s1=pd.Series([1,2,3,4])
s2=pd.Series([1,2,3,4,5,6,7])
print(s1,s2)

```

```python
print(s1+s2) # for extra rows we will get Nan
```

# DataFrame
## 2d data 


```python
#from list 

l=[1,2,3,4,5,6]
df=pd.DataFrame(l)
print(df)
print(type(df))
```

```python
# using dictionary
d={'a':[1,1,1,1],'b':[2,2,2,2],'c':[3,3,3,3]}
```

```python
df=pd.DataFrame(d,columns=["a","b"],index=['a','s','c','d']) # displays only a,b column
```

```python
print(df)
```

```python
print(df['b']['c']) # df[column][row]
```

```python
list_1=[[1,2,3,4],[2,3,4,5]]
df_2=pd.DataFrame(list_1)
print(df_2)
```

```python
dic={'s':pd.Series([1,2,3,4]),'r':pd.Series([2,3,4,5])}
df_2=pd.DataFrame(dic)
print(df_2)
```
