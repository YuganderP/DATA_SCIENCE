# Arthematic Operations in pandas

```python
# we will perform arthematic operations on the column data 
dic={'a':[1,2,3,4],'b':[3,4,5,6]}
df_3=pd.DataFrame(dic)
```

```python
print(df_3)
```

```python
# df_3['c']=df_3['a']+df['b']
# print(df_3)

df_3['c']=(df_3['a']+df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']-df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']*df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']/df_3['b'])
print(df_3)
print()
df_3['c']=(df_3['a']//df_3['b'])
print(df_3)
print()
```

```python
df_3['c']=((df_3['a']<3) & (df_3['b']<5))
print(df_3)
print()
```

# merge and concat

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,33,8]})
var2=pd.DataFrame({'A':[1,2,3,5],'B':[5,6,7,8]})
# print(pd.merge(A,B))
# merging common port 

print(pd.merge(var1,var2,on='A'))
print()
print()
# pd.merge(var1,var2,how="right")
print()
pd.merge(var1,var2,how="outer",indicator=True)
print()

pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","id")) # displays all data 

```

# concat

```python
#to join 2 series

sr1=pd.Series([1,2,3,4])
sr2=pd.Series([5,6,7,8])

sr1
print()
sr2

```

```python
pd.concat([sr1,sr2])
```

```python
d1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
d2=pd.DataFrame({'A':[1,2,3,5],'C':[5,6,7,8]})


pd.concat([d1,d2])
```

```python
pd.concat([d1,d2],axis=1) # column merger
```

```python
d1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
d2=pd.DataFrame({'A':[1,2],'C':[9,18]})
```

```python
pd.concat([d1,d2],axis=1,join="inner")
```

```python
d1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
d2=pd.DataFrame({'A':[1,2,3,4],'C':[9,18,10,11]})

pd.concat([d1,d2],keys=["d1","d2"])
```

# join and Append

```python
d1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]},index=['a','b','c','d'])
d2=pd.DataFrame({'C':[1,32,5,4],'D':[9,18,10,11]},index=['a','b','c','d'])
```

```python

d1.join(d2,how="outer")
```

# Append

```python
d1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
d2=pd.DataFrame({'B':[1,2],'C':[5,6]})
pd.concat([d1,d2])
```
