# 1. Introduction to Pandas

Pandas is one of the most important Python libraries for:

- Data Analysis
- Data Cleaning
- Data Manipulation
- Exploratory Data Analysis (EDA)

Main Data Structures:

1. Series → Single column
2. DataFrame → Complete table


```

import pandas as pd
```


# 2. Series

A Series is a one-dimensional labeled array.

Think of it like a single column in Excel.


```

s = pd.Series([10, 20, 30, 40])
print(s)
```


Explanation:

- Left side numbers are indexes
- Right side numbers are values



# 3. DataFrames

A DataFrame is a table with rows and columns.

This is the most commonly used Pandas structure.


```

data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [23, 25, 22],
    "Marks": [90, 85, 95]
}

df = pd.DataFrame(data)
print(df)
```


Explanation:

- Each column represents a feature
- Each row represents one record


# 5. Inspecting Data


```

print(df.head())
print(df.info())
```


`head()` shows the first few rows.

`info()` gives:
- column names
- data types
- missing values



# 6. Selecting Columns


```

print(df["Name"])
```


This selects only the `Name` column.



# 7. loc vs iloc

This is one of the MOST IMPORTANT Pandas concepts.


```

print(df.loc[0])
```


`loc` uses labels/index names.


```

print(df.iloc[0])
```


`iloc` uses positions.



# 8. Filtering Data

Filtering means selecting rows that match conditions.


```

print(df[df["Marks"] > 90])
```


This means:

Show only rows where marks are greater than 90.



# 9. Adding Columns


```

df["Passed"] = True
print(df)
```


We created a new column called `Passed`.



# 10. Missing Values


```

print(df.isnull())
print(df.isnull().sum())
```


Real-world datasets often contain missing values.



# 11. Sorting Data


```

print(df.sort_values("Marks", ascending=False))
```


This sorts students from highest marks to lowest.



# 12. Aggregation Functions


```

print(df["Marks"].mean())
print(df["Marks"].max())
print(df["Marks"].min())
```


Aggregation summarizes data.



# 13. GroupBy

GroupBy is one of the most powerful Pandas concepts.


```

data = {
    "Department": ["IT", "HR", "IT", "HR"],
    "Salary": [50000, 40000, 60000, 45000]
}

emp = pd.DataFrame(data)

print(emp.groupby("Department")["Salary"].mean())
```


This groups employees department-wise and calculates average salary.



# 14. String Operations


```

print(df["Name"].str.upper())
```


Converts all names to uppercase.



# 15. Apply Functions


```

def grade(score):
    if score >= 90:
        return "A"
    return "B"

df["Grade"] = df["Marks"].apply(grade)

print(df)
```


`apply()` lets us use custom logic on columns.



# 16. Merge DataFrames


```

df1 = pd.DataFrame({
    "ID": [1,2],
    "Name": ["Alice", "Bob"]
})

df2 = pd.DataFrame({
    "ID": [1,2],
    "Marks": [90,85]
})

merged = pd.merge(df1, df2, on="ID")

print(merged)
```


Merging combines datasets together.



# 17. Pivot Tables


```

sales = pd.DataFrame({
    "Region": ["East", "East", "West"],
    "Year": [2024, 2025, 2024],
    "Sales": [100, 200, 300]
})

pivot = sales.pivot_table(
    values="Sales",
    index="Region",
    columns="Year",
    aggfunc="sum"
)

print(pivot)
```


Pivot tables summarize data in a report-like format.



# 18. Performance Tip

Use vectorized operations instead of loops whenever possible.


```

df["DoubleMarks"] = df["Marks"] * 2
print(df)
```


This is much faster than using loops.
