```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y) # creates a line graph
plt.title("sample graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
# subplots in matplotlib
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**y,y)
plt.subplot(2,2,4)
plt.plot(x**0.24,y)
```

```python
# object oriented way to make changes to canvas we can change dpi and figure size
fig=plt.figure(figsize=(6,8),dpi=100) # 600X800
axis1=fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.7,0.2,0.2,0.2])
axis2.plot(y,x)
axis2.set_title("second Graph")
```

## box plot and voiln plot

these 2 plots to show the distribution of the categorical data boxplot( box and whisker plot) shows the ditribution of quantitative data in a way
that facilitates comparision b/w variables or across levels of categorical variable The box shows quartiles of the dataset while the whiskers exted to show
the rest of the distribution except for the points that are determined to be outliers using a method that is a function of the inter-quartile range

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

# strip plot / swarm plot

strip plot displays individual data points for each category
points may overlap when values are close

swarm plot:
Displys individual datapoints for each category without overlapping making the distribution easy to understand

```python
sns.stripplot(x='day',y='total_bill',data=df)
```

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

```python
sns.violinplot(x='tip',y='day', data=df)
sns.swarmplot(x='tip',y='day',data=df)
```
