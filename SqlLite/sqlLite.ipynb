```python
import sqlite3
connection=sqlite3.connect('example.db')
```

```python
connection
```

```python
cursor=connection.cursor()
```

```python
# CREATING A  table 

cursor.execute('''
Create Table If Not Exists employees(
id Integer Primary Key,
name text Not Null,
age Integer,
department text
) 
''')
```

```python
connection.commit()
```

```python
cursor.execute(''' select * from employees ''')
```

```python
# inserting the data 

cursor.execute(''' 
Insert Into employees(name,age,department) values('yug',29,'Finance')
''')
```

```python
# query the data 

cursor.execute('select * from employees')
rows=cursor.fetchall()
print(rows)
```
