```python
#filing the null records
# filling with mean values
updated_df=df
updated_df['Age']=updated_df['Age'].fillna(df['Age'].mean())
updated_df.info()
```

```python
# filling with median 

updated_df1=df
updated_df1['Age']=updated_df['Age'].fillna(df['Age'].median())
updated_df1.info()
```
